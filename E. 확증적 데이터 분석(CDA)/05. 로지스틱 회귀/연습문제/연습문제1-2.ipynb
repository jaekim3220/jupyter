# 문제2

다음의 데이터는 타이타닉 탑승객 명단 데이터이다.

적절한 데이터 전처리와 정제를 수행한 후 분석하라.

단, 모든 독립변수는 정규분포를 만족한다고 가정한다.

```python
df = read_excel("https://data.hossam.kr/E05/titanic.xlsx")
df.head()
```

```python
df.info()
```

```python
df1 = df.copy()
df1.isnull().sum()
```

```python
print(df1.columns)
```

```python
df2 = df1.drop(columns=['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin'], axis=1)
df2.fillna(df1['Age'].median(skipna=True), inplace=True)
df2.isna().sum()
```

```python
df2
```

```python
df2['Sex'].value_counts()
df2['Embarked'].value_counts()
```

```python
df3 = df2.copy()
df3['Sex'] = df2['Sex'].map({'male':0, 'female':1})
df3.head()
```

```python
df3['Embarked'] = df3['Embarked'].map({'S':'0', 'C':'1', 'Q':'2'})
df3.head()
```
