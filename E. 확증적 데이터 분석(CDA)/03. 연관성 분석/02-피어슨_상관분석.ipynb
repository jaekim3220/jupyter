## #03. 공분산

공분산은 `상관관계의 상승 혹은 하강하는 경향을 이해`할 수는 있으나 2개 변수의 측정 단위의 크기에 따라 값이 달라지므로 `절대적 정도를 파악하기에는 한계`가 있다.

결과값이 양수면 정비례, 음수면 반비례

>그래서 잘 안쓴다.

```python
np.cov(df['언어영역'], df['수리영역'])[0, 1]
```

## #04. 상관분석

### numpy를 통한 상관계수 구하기

`p-value`는 알 수 없기 때문에 분석 결과를 검정할 수 없다.

| | 언어 | 수리 |
| -- | -- | -- |
| 언어 | 1 | 0.82106421 |
| 수리 | 0.82106421 | 1 |

```python
np.corrcoef(df['언어영역'], df['수리영역'])
```

```python
np.corrcoef(df['언어영역'], df['수리영역'])[0,1]    #0행 1열이 상관계수, 1행0열이 상관계수
```

### 피어슨 상관계수와 `p-value` 구하기

```python
r,p = stats.pearsonr(df['언어영역'],df['수리영역'])
"상관계수 : %f, p-value : %f"%(r,p)
```

#### 결과 해석

결과의 `상관계수`의 값이 `0.821`, `p-value(유의수준)`값이 `0.024`로 0.05보다 작으므로 `귀무가설을 기각`하고 `대립가설을 채택`한다. 그러므로 언어영역 점수와 수리영역 점수는 상관이 있다.

### 데이터프레임 객체를 통한 상관계수 구하기

numpy를 통해 구한 결과와 동일하다.

이와 같이 표시되는 형태를 상관행렬이라고 한다.

```python
corr = df.corr(method='pearson')
corr
```

### 상관분석 결과 시각화

heatmap

```python
plt.rcParams['figure.figsize']=(5,3)

sb.heatmap(corr,
           annot=True,  #실제 값 화면에 표시
           cmap="Greens",    #색깔
           vmin=-1, vmax=1, #컬러차트 영역(상관계수 범위:-1~1)
           linewidths=.5    #구분선
           )
plt.show()
plt.close()
```

## 예제 (2) - mtcars 데이터를 통한 실린더수와 연비의 상관관계 분석

R 패키지인 ggplot2에 내장된 샘플 데이터로 자동차 32종의 11개 속성에 대한 정보를 담고 있는 데이터

| 필드 | 설명 |
|---|---|
| mpg | 연료 효율성 (마일 당 갤런) |
| cyl | 실린더 수 |
| disp | 배기량 (cubic inches) |
| hp | 마력 (horsepower) |
| drat | 후륜축 비율 (rear axle ratio) |
| wt | 차량 무게 (1000 파운드) |
| qsec | 1/4 마일 달성 시간 (초) |
| vs | V/S (V 엔진 형태인지, S는 직렬 엔진 형태인지를 나타냄) |
| am | 변속기 형태 (0: 자동, 1: 수동) |
| gear | 전진 기어 수 (전송 기어) |
| carb | 기화기 개수 (carburetors) |

### 데이터 가져오기

```python
df = read_excel("https://data.hossam.kr/E03/mtcars.xlsx", index_col='name')
df
```

### 실린더 수에 따른 연비에 대한 산점도 그래프

```python
df.columns
```

```python
plt.rcParams['figure.figsize']=(10,5)
sb.regplot(data=df, x='cyl',y='mpg')
plt.grid()
plt.show()
plt.close()
```

#### 중간 해석

실린더 수가 증가함에 따라 연비가 낮아지는 경향이 있는것으로 보임

### 상관분석

```python
plt.rcParams['figure.figsize']=(5,3)
corr = df.filter(['cyl', 'mpg']).corr(method='pearson')

sb.heatmap(corr,
           annot = True, #실제 값 화면에 나타내기
           cmap = 'Reds', #색상
           vmin = -1, vmax=1 , #컬러차트 영역 -1 ~ +1
           linewidths=.5 #구분선
           )
plt.show()
plt.close()
```

#### 중간 해석
분석결과 상관계수가 -0.85로 나타나 실린더수와 연비는 강한 음의 상관관계가 있는 것으로 나타남

### 분석결과 검정

피어슨 상관계수와 `p-value` 구하기

```python
stats.pearsonr(df['cyl'],df['mpg'])
```

### 결과보고

32종의 자동차를 대상으로 실린더수에 따른 연비를 조사한 결과 실린더 수와 연비의 상관계수($r$)는 `-0.852`으로 강한 `음의 상관관계`를 갖는 것으로 타나났다. `p-value(유의수준)`값이 `6.112e-10`로 유의수준 0.05보다 작으므로 이는 통계적으로 유의하다.
