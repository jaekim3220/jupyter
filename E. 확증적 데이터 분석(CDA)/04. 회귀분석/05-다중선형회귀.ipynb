```python
df = read_excel("https://data.hossam.kr/E04/boston.xlsx")
df.head()
```

## 파이썬의 ols 객체로 분석

```python
print(df.columns)
```

```python
model = ols("MEDV ~ CRIM + INDUS", data=df)
fit = model.fit()
fit.summary()
```

R-squared:	0.278로 적합성이 매우 떨어지기 때문에 가공이 필요
