# 문제1

x축의 간격을 10일 간격으로 설정하여 날짜별 종가와 시가 변화를 구현

```python
df = read_excel("https://data.hossam.kr/bitcoin.xlsx")
df
```

X축 설정

```python
xlabels = list(df['날짜'])
xlabels
```

x축 범위 설정

```python
x = np.array(range(0, len(xlabels), 10))
x
len(x)
```

y축 데이터 설정

```python
ylabels1 = list(df['시가'])
ylabels1
```

```python
y1 = ylabels1[::10]
y1
len(y1)
```

```python
ylabels2 = list(df['종가'])
y2 = ylabels2[::10]
len(y2)
```

다중 막대 그래프 그리기

```python
plt.figure()    # 그래프 시작을 선언
plt.title("10일 간격의 시가와 종가 ")    # 그래프 제목 설정
# 서로 다른 y 값을 갖는 두 종류의 막대를 하나의 그래프에 생성
plt.bar(x-2, y1, color = 'orange', label = '시가', width = 4)   
plt.bar(x+2, y2, color = 'blue', label = '종가', width = 4)
plt.xlabel('날짜')  # x축 제목
plt.ylabel('시가와 종가')  # y축 제목
plt.grid()  # 그리드 생성
plt.legend()    # 범례 표시(위치 자동-수동 가능)
# plt.xticks(x, xlabels)  # x축 변수 변경
plt.show()
plt.close()
```

# 문제4

요구사항 1\
종사자 규모에 따른 업체 수를 확인할 수 있는 세로 막대 그래프를 구현하시오. 이 때 각 데이터의 범주는 년도로 설정

요구사항 2\
종사자 규모별 업체 수의 비율을 시각화 하시오. 2015년도와 2016년도에 대한 시각화 결과물이 각각(총 2개) 구현

```python
df = read_excel("https://data.hossam.kr/jobs_by_scale.xlsx")
df
```

```python
xlabels = list(df['종사자규모별'])
xlabels
```

```python
x = np.array(range(0, len(xlabels)))
x
```

```python
y1 = list(df[2015])
y1
```

```python
y2 = list(df[2016])
y2
```

```python
plt.figure()
plt.title("종사자 규모에 따른 업체 수")
plt.bar(x-0.2,y1,color='#0066ff', label="2015년", width=0.4)
plt.bar(x+0.2,y2,color='#ff6600', label="2016년", width=0.4)
plt.xticks(x, xlabels)  # x축 변수 변경
plt.xlabel("종사자규모")
plt.ylabel("업체 수")
plt.grid()  # 그리드 생성
plt.legend()    # 범례 표시(위치 자동-수동 가능)
plt.show()
plt.close()
```

# 문제 5

교통사고 발생건수와 부상자수의 상관관계를 알아보고자 한다. 적합한 시각화 결과물을 제시

```python
# Scatter Plot
df = read_excel("https://data.hossam.kr/traffic_acc.xlsx")
df
```

```python
발생건수 = list(df["발생건수"])
# 발생건수
사망자수 = list(df["사망자수"])
# 사망자수
```

```python
plt.figure()    # 그래프 시작
plt.title("교통사고 발생건수와 부상자수의 상관관계")    # 그래프 제목
plt.xlabel("교통사고 발생건수")
plt.ylabel("사망자수")
plt.scatter(발생건수, 사망자수) # 그래프 생성
# plt.legend()
plt.grid()  # 그리드 표시
plt.show()  # 그래프 시각화
plt.close() # 그래프 종료
```
