```python
from pandas import DataFrame, merge, concat, Series, read_excel, pivot_table
```

문제1

```python
a1 = DataFrame({'A1':[62, 60, 63, 59, 61]},
               index=['1회', '2회', '3회', '4회', '5회']
                  )
a1
```

```python
a2 = DataFrame({'A2':[62, 67, 71, 64, 65, 66]},
               index=['1회', '2회', '3회', '4회', '5회', '6회']
                  )
a2
```

```python
a3 = DataFrame({'A3':[68, 66, 71, 67, 68, 69, 70]},
               index=['1회', '2회', '3회', '4회', '5회', '6회', '7회']
                  )
a3
```

```python
a4 = DataFrame({'A4':[58, 62, 60, 59, 61, 63, 64]},
               index=['1회', '2회', '3회', '4회', '5회', '6회', '7회']
                  )
a4
```

```python
merge1 = merge(a1, a2, left_index=True, right_index=True, how='outer')
merge1
```

```python
merge2 = merge(merge1, a3, left_index=True, right_index=True, how='outer')
merge2
```

```python
finalDataFrame = merge(merge2, a4, left_index=True, right_index=True, how='outer')
finalDataFrame
```
